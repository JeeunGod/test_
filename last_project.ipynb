```python
import pandas as pd
from numpy import NaN
import matplotlib.pyplot as plt

operator = pd.read_csv('operators.csv')
fig = plt.figure(figsize=(55,25))
axis = fig.add_subplot(1, 1, 1)
oper = pd.melt(operator, id_vars = ['Name', 'Role', 'Primary1', 'Secondary1', 'Gadget1'])
op = oper[['Name', 'Role', 'Primary1', 'Secondary1', 'Gadget1']]
axis.scatter(op['Name'], op['Primary1'])
axis.set_xlabel('Operators Name')
axis.set_ylabel('Primary Guns')
axis.set_title('Primary Guns by Operators')   #오퍼레이터별 선호하는 총기를 산점도 그래프로 표현
```

```python
oper.groupby('Gadget1')['Name'].unique() #각 오퍼레이터 별 특수무기에서 선택율이 높은 gadget
```
